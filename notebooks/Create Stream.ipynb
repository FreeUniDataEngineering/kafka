```python
import csv

from time import sleep


input_data = 'data/character_stats.csv'
output_path = 'individual_heroes'

 
file = open(input_data, 'r')

next(file) # to skip header
records = file.readlines()
 
count = 1    
for line in records:
    print(f'Processing Hero # {count}')
    words = line.strip().split(',')
    hero_name = words[0]
    print(line)
    
    with open(f'{output_path}/{hero_name}.csv', 'w') as f:
        writer = csv.writer(f)

        writer.writerow([*words, str(count)])
    
    count += 1
    sleep(5)
```
